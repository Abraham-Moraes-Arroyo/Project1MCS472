# Exercise 2 Lecture 8

Let H(z) be the transfer function of a filter to remove a signal at 60Hz, sampled at 720 Hz. 

1. Make the Bode plot for this filter. 

2. Use the Bode plot to explain which signals are also removed by this filter. 

Based on bodeplots.ipynb 

```julia
T= 1/720
```

We let the frequencies range from 0 to 720. To make the plot, we evaluate the formulas $r$ and $\phi$.

```julia
freq = 0:60:720
omega = [2*pi*n for n in freq]
rw = [sqrt(1 + cos(w*T))/sqrt(2) for w in omega]
phi = [-atan(sin(w*T)/(1+cos(w*T))) for w in omega]
plot(phi, rw, proj=:polar, label="(1+z^(-1))/2")
```

```julia
plot(freq, rw, label="amplitude gain", legend=:bottomright, xlabel="frequency Hz")
```

```julia
plot(freq, phi, label="phase shift", legend=:topright, xlabel="frequency Hz")
```

```julia
gainplot = plot(freq, rw, label="gain", legend=:top, xlabel="frequency Hz")
phaseplot = plot(freq, phi, label="phase", legend=:topright, xlabel="frequency Hz")
plot(gainplot,phaseplot)
```

The signals that are going to be removed are those who are divisible by 60. 

# Exercise 4 L8

Show that an unstable plant with transfer function. 

$$
    P(z)=\frac{1}{1-\alpha Z^{-1}}, |\alpha| > 1,
$$

is stabilized by proportional feedback with transfer function

$$
    H(z)= \beta
$$

# Solution: 
The reason to why this is an unstable is because of $z^{-1}$. 

An example on how to stablize the transfer function: (Seen in L-8, slide 25/26)

The transfer function of the closed loop is then

$$
    \frac{P}{1+PH}= \frac{1- \frac{2}{z}}{1 + \frac{3}{2}(1- \frac{2}{z})} = \frac{1}{1+\frac{2}{z} +\frac{3}{2}}= \frac{1}{\frac{5}{2}- \frac{2}{z}}
$$

$$
    = \frac{\frac{2}{5}}{1- \frac{4}{5}Z^{-1}}
$$

$$
    = \frac{2}{5}(1 +\frac{4}{5}z^{-1}+(\frac{4}{5})z^{-2}+\ldots)
$$

As $k \rightarrow \infty$, $(\frac{4}{5})^{k} \rightarrow$ 0, so the closed loop is stable.
